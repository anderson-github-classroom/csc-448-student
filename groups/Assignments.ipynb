## Weekly Group Assignments
Please use the specific Zoom link each week that is listed by your name. The link will change every week in an effort to help cross-pollinate the course. 

```python
import pandas as pd
zoom_link = "https://calpoly.zoom.us/j/95497024735?pwd=OUlMcXUrd1RXRHBpa21QcHZaZXU2UT09"
breakout_rooms = pd.Series([
    "A",
    "B",
    "C",
    "D",
    "E"
],index=["Group 1","Group 2","Group 3","Group 4","Group 5"])
breakout_rooms
```

```python
import numpy as np

students = pd.read_table("students.tsv",sep="\t")

def make_clickable(val):
    # target _blank to open new window
    return '<a target="_blank" href="{}">{}</a>'.format(zoom_link, val)

weeks = {}
for i in range(1,10):
    col = 'Week %d'%i
    students[col] = np.random.choice(breakout_rooms.index,size=len(students))
    weeks[col] = students[["Email",col]]
    weeks[col] = weeks[col].style.format({col:make_clickable})
```

### Week 1
Recommendations of outline for group meeting (guidelines only and you should adapt as needed). Feel free to share screens when it makes sense.

5. Everyone share an update on what they've done on the project recently. Find anything interesting? What are your biggest questions?
6. What are you going to try on the project next?

What if you get through all of this and still have time remaining? Or more likely you still have time remaining in the lab portion. Then I suggest to focus on the project. The other altnerative that I would suggest is let folks discuss anything COVID-19 related with a spin towards the project.
